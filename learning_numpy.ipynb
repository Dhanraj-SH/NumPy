#### The basics

The numpy's main objective is the homogenous multidimensional array. It is a table of elements(usually number) all of the same type, indexed by a tuple of non-negative integer. In NumPy dimensions are called axes.

Lets take an example where a array coordinates of a point in 3D space [1, 2, 1] has one axis. Tha axis has 3 elements in it, so we say it has a lenght of 3.

[[1., 0., 0.],
 [0., 1., 2.]]

for the above array which is a 2d array which means the array as 2 axes 

`Note:` axes = dimensions of array

- axes 0 = 2(2 rows)
- axes 1 = 3(3 columns)

NumPy array class is called `ndarray`. It is also known by alias array. and `numpy.array` is not same as the standard python library class `array.array`, which only handles one dimensional arrays and offers less functionality. 

The imp attributes of `ndarray` objects are:

`ndarray.ndim` - returns the number of axes(dimensions) of the array.

`ndarray.shape` - the dimensions of the array. This is a tuple of integers indicating the size of the array in each dimensions. For a matrix with n rows and m colomns, the shape will be(m,n). The length of the `shape` tuple is therefore the number of axes, `ndim`

`ndarray.size` - it returns the total number of elements in the array which is equal to the product of the `shape` (m * n)

`ndarray.dtype` - An object describing the type of element in the array. Where on can create or specify dtype's using the standard python types

`ndarray.itemsize` - The size in of each elements in bytes which is present in array. For example, an array of elements of type `float64` has `itemsize` 8(=64/8), while one of type `complex32` has `itemsize` 4(=32/8). It is equivalent to `ndarray.dtype.itemsize`.

`ndarray.data` - The buffer containing the actual elements of the array. Normally we won't need to use this attribute because we will access the elements in an array using indexing facilities.

#### Array creation 

There are several ways to create an array from a regular python list or tuple using `array` function. The of resulting array is deduced from the type of the elements in the sequences.


```python
a = np.array([1, 3, 5])

a.dtype
```

```python
b = np.array([1.2, 3.4, 5.6])

b.dtype
```

```python
c = np.array([[1, 2], [3, 4]], dtype = complex) # the array dtype can also be explicity specified at the creation time

c
```

Often, the elments of array is unknown so the numpy offers serveral functions to create initial placeholders content. These minimize the necessity of growing arrays, an expensive operation.

The function `zeros` creates an array full of zeros, and the function `ones` creates a function full of ones and the `empty` functions creates an array whose initial content is random and depends on the state of the memory. By default, the dtype is created as float64 but can be specified using `dtype` argument

To create a sequence of numbers, NumPy provides the `arange` function which is analogous to the python built-in `range`, but returns an array.

```python
np.zeros((3,4))
```

```python
np.ones((3, 2), dtype = np.int32)
```

```python
np.empty((2, 3))
```

```python
np.arange(10, 30, 5) # np.arange(start, end, increment)
```

```python
np.arange(15) # by defalut starts from zero
```

```python
np.arange(4, 14) # starts from 4 and ends at 13
```

```python
np.arange(0, 3, 0.3) # also accepts float values
```

When `arange` is used with floating point arguments, it is generally not possible to predict the number of elements obtained, due to the finite floting point percision. For this reason , it is usaully better to use the function `linspace` that recives as an argument the number of elements that we want, instead of the step

```python
from numpy import pi

np.linspace(0, 2, 9) # 9 numbers from 0 to 2
```

```python
x = np.linspace(0, 2*pi, 100)
f = np.sin(x)

x
```
